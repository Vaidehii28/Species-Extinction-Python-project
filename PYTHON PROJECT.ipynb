```python
class Species:
    def __init__(self, spName, spCount, extnct, thrtnd):
        self.specieName = spName
        self.count = spCount
        self.extinct = extnct
        self.threatened = thrtnd

    def showData(self):
        print("--"+self.specieName+"--")
        print("-> There are about " + self.count + " " + self.specieName + " species identified till date!")
        print("-> Number of " + self.specieName + " species that have gone extinct is: " + self.extinct)
        print("-> Number of species threatened with extinction are: " + self.threatened)
    def prevention(self):
        print("\n------------------------------------------------------------")
        print("Ways to prevent the  extinction of the species: ")
        print(" 1. Buy eco-friendly products\n 2. Eat less meat\n 3. Dont buy souvenirs made from endangered species \n 4. Follow 3r rules \n 5. Spread awareness and also get involved \n")



class Birds(Species):
    def __init__(self):
        Species.__init__(self, "Bird", "11,158", "159", "1,481")
        Species.showData(self)
        Species.prevention(self)



class Mammals(Species):
    def __init__(self):
        Species.__init__(self, "Mammal", "6,458", "85", "1,317")
        Species.showData(self)
        Species.prevention(self)


class Fishes(Species):
    def __init__(self):
        Species.__init__(self, "Fish", "35,672", "80", "3,040")
        Species.showData(self)
        Species.prevention(self)


class Amphibians(Species):
    def __init__(self):
        Species.__init__(self, "8,250", "11158", "35", "2,390")
        Species.showData(self)
        Species.prevention(self)


class Reptiles(Species):
    def __init__(self):
        Species.__init__(self, "Reptile", "11,341", "30", "1,449")
        Species.showData(self)
        Species.prevention(self)


class Crustaceans(Species):
    def __init__(self):
        Species.__init__(self, "Crustacean", "80,122", "11", "742")
        Species.showData(self)
        Species.prevention(self)


def printChoice(arg):
    if arg == 1:
        Birds()
    elif arg == 2:
        Mammals()
    elif arg == 3:
        Fishes()
    elif arg == 4:
        Amphibians()
    elif arg == 5:
        Reptiles()
    elif arg == 6:
        Crustaceans()
    elif arg == 7:
        print("Thanks!")
    else:
        print("You entered wrong preference!")


if __name__ == "__main__":
    print("Quick Data!")
    print("------------------------------------------------------------")
    i= int(input("Enter your preference: \n1.Bird\n2.Mammals \n3.Fishes\n4.Amphibians\n5.Reptiles\n6.Crustaceans\n7.Exit\n"))
    printChoice(i)
```
